# Chapter 1: What is quantum advantage?

## 1.1 Introduction

Quantum computing holds the promise of solving problems that are intractable for classical computers, enabling new frontiers in chemistry, materials discovery, optimization, and beyond. Recent advances in quantum hardware and software have brought us closer to realizing this potential, with quantum systems now capable of executing increasingly complex algorithms with improved fidelity. A central goal in the field is to reach **quantum advantage**: the point at which a quantum computer performs a useful task better than a classical one. This does not mean quantum computers will replace classical high-performance computing systems. More likely, they will be used alongside them, with each handling the parts of a problem they are best suited for.


`Criterion_1` is critical since quantum computers solve problems not efficiently solvable by classical methods, making direct comparison challenging. There are three main ways to address this:

1. **Error bars and error bounds**: Ensuring every computation step was carried out reliably; for example, through fault-tolerant quantum computing, rigorously justified error mitigation, or post-selected error detection.

2. **Problems with efficient classical verification**: Certain problems can only be solved efficiently on quantum computers, but their solutions can be verified efficiently classically (e.g., sampling problems).

3. **Variational problems**: In these cases, the solutions cannot be directly verified classically, but can be scored and compared with classical approaches. For example, in ground-state energy estimation, variational methods provide upper bounds that can be compared to find the best solution.

`Criterion_2` refers to demonstrable performance advantages over classical methods. Strictly speaking, unconditional algorithmic separations between quantum and classical computing have not been proven in general, though this is considered extremely unlikely. All currently known separations depend either on the best known classical algorithms (e.g., Shor's factoring algorithm) or on complexity theory assumptions. For this reason, any claim of quantum separation must be tested against both current classical methods and future improvements in classical algorithms.

The path to quantum advantage focuses on three key problem families: sampling problems (generating probability distributions), variational principle problems (finding ground states or optimal solutions), and expectation values of observables (measuring physical properties of quantum systems). Each offers a possible route to quantum advantage, since the results can be validated and compared fairly against classical approaches.

## 1.3 Problems Poised for Quantum Advantage

Quantum simulation in chemistry and materials science is often seen as one of the strongest candidates for near-term quantum advantage. These problems are difficult for classical computers because the cost of simulating quantum systems grows very quickly with system size.

On the other hand, combinatorial optimization problems, particularly those tackled using the Quantum Approximate Optimization Algorithm (QAOA), also show significant potential for quantum advantage. 

Rather than seeking general-purpose speedups across all problem types, a viable strategy is to focus on targeting specific tasks where quantum computers can provide demonstrable value, building a portfolio of applications where quantum advantage can be rigorously established and verified.

## 1.4 Elements for Advantage

Achieving quantum advantage requires the integration of three essential error-management strategies: Quantum Error Correction (QEC), which protects quantum information from decoherence; Quantum Error Mitigation (QEM), which reduces the impact of noise in near-term devices; and Error Detection, which identifies and handles faulty operations. These techniques must work together with high-performance quantum hardware, such as superconducting qubits with tunable couplers for improved gate fidelities and dynamic circuits that enable mid-circuit measurements and conditional operations.

The path to advantage also depends on Quantum-Centric Supercomputing (QCSC) architectures, in which quantum processing units (QPUs) complement traditional CPUs and GPUs in hybrid workflows, with each processor handling the tasks it is best suited for. An important part of this ecosystem is the Qiskit Runtime environment, which acts as the orchestration layer between quantum hardware and classical computing resources. Through Qiskit Runtime, researchers can run hybrid algorithms, apply error-mitigation techniques, and coordinate data flow between quantum and classical processors within a common framework, while still retaining detailed control when needed.

## What is the Quantum Advantage Tracker?

The [Quantum Advantage Tracker](https://quantum-advantage-tracker.github.io/) is a community-driven platform that documents experimental demonstrations of quantum advantage. The tracker organizes quantum advantage demonstrations into three categories:

### Finding the Ground State Energy of Fe₄S₄

The [Quantum Advantage Tracker](https://quantum-advantage-tracker.github.io/) documents quantum computing achievements that aim to demonstrate advantage over classical methods. The entry we are interested in focuses on computing the ground state energy of $ \mathrm{Fe}_4\mathrm{S}_4$ (iron-sulfur cluster), a biologically important molecule found in proteins. This is a challenging problem because transition metal systems have strong electron correlations that are difficult to simulate classically.

The quantum computation uses the TrimCI (Trimmed Configuration Interaction) method from the paper ["From Random Determinants to the Ground State"](https://arxiv.org/pdf/2511.14734). The approach builds accurate ground states from random Slater determinants without requiring pre-selected reference states. It uses quantum circuits to prepare multi-determinant states, then systematically trims away unimportant components. Remarkably, for the 8 × 8 Hubbard model it recovers over 99% of the ground-state energy using only $10^{-28}$ of the Hilbert space, surpassing classical methods like AFQMC.
